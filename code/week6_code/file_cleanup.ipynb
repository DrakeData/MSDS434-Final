```python
import pandas as pd
import numpy as np
import os
```

```python
# Read in file
main_df = pd.read_csv('spotify_tracks.csv')
main_df.head()
```

```python
len(main_df)
```

```python
# create new column for bop or flop
main_df['bop_or_flop'] = np.where(main_df['popularity'] >= 70, 1, 0)
```

```python
main_df.head()
```

```python
test_df = main_df.sample(70000, replace=True)
len(test_df)
```

```python
test_df.head()
```

```python
test_df['bop_or_flop'].value_counts()
```

```python
test_df.to_csv('spotify_train.csv')
```

```python
test_id_ls = test_df['id'].unique().tolist()
test_id_ls[:5]
```

```python
sample_df = main_df.sample(70000, replace=True)
sample_df = sample_df[~sample_df.id.isin(test_id_ls)]
len(sample_df)
```

```python
sample_df['bop_or_flop'].value_counts()
```

```python
sample_df.to_csv('spotify_test.csv')
```
